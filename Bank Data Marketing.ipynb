## Need of correlation

Correlation is defined as the linear relationship between two variables ,how one variable depends on another variable


correlation is of three types:|
    
1.Positive correlation-in this type of correlation in which two variables move in same direction,if x increases ,y also increases

2.Negative correlation-in this type of correlation in which two variable moves in opposite direction,if x increases why decreases

3.zero correlation - when there is no relationship between the 2 variables

A perfect negative correlation is represented by the value -1.0,

 0 indicates no correlation,

+1.0 indicates a perfect positive correlation.



### How to Find Correlation

```python
df_test_oversample.corr()# this is the theoritical way to find correlation between variables
```

# Logistic Regression

Logistic Regression is used when the dependent variable(target) is categorical in nature.

```python
df_test_oversample.columns
```

```python
x=df_test_oversample[['age', 'job', 'marital', 'education', 'balance', 'contact', 'day',
       'month', 'duration', 'campaign', 'previous', 'poutcome', 'housing_new',
       'loan_new']]
y=df_test_oversample["Deposit_new"]
```

```python
y.value_counts() # to see values in y variable
```

```python
x # to see values in x variable
```

#### Split the data into training set and testing set using train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101) # Split the data into training set and testing set using train_test_split
```

```python
y_test.value_counts()
```

```python
x_test
```

#### Training the Model

```python
from sklearn.linear_model import LogisticRegression # Importing linear regression model
```

```python
lm=LogisticRegression() # lm depicts the logmodel we are using 
```

```python
lm.fit(x_train,y_train)
```

#### Predictions and Evaluations

```python
predictions=lm.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

True Positive=9799,it means clients who have deposited in scheme and machine is predicting correctly

False Positive= 2122,it means client has  not deposited  in scheme but machine is showing he has deposited ,also caled Type I error

False Negative=2437,it means client has actually deposited in scheme but machine is showing he has  not deposited ,also called Type II error

True Negative=9595,it means client has not deposit in schem and machine is also showing he has not deposited

```python
print(classification_report(y_test,predictions))
```

**Accuracy =0.81-> ratio of correctly classify clients who has deposited to total no.of clients**



(TP+TN)/(TP+TN_FP_FN)

**Precision=0.81-> ratio of correcly classified clients to the total no.of clients**

(TP)/(TP+FP)

**Recall=0.81-> is the ratio of machine classified clients who deposited to actual clients who has deposited**

(TP)/(TP+FN)

**F1-Score=0.81->is the equibilirium between recall and precision**

2(precision*recall)/(precision +recall)
